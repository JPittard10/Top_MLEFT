```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load posterior
with open("limit_results/truth_test_fit/ctGRe_ctu8/posterior.json") as f:
    data = json.load(f)

# Example: pick two coefficients
c1 = "ctGRe"
c2 = "ctu8"   # change to your actual names

x = np.array(data[c1])
y = np.array(data[c2])

# Make contour plot
plt.figure(figsize=(6,6))

sns.kdeplot(
    x=x,
    y=y,
    levels=[0.68, 0.95],   # confidence levels
    fill=True,
    alpha=0.4
)

# Optional: mark SM point (0,0)
plt.scatter(0, 0, color='black', marker='*', label='SM')

plt.xlabel(r"$c_{tG}$")
plt.ylabel(r"$c_{tu}$")
plt.legend()
plt.grid(True)

plt.show()
```
